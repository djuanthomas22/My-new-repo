```python
import random

# Ask: roll the dice?
roll_count = 0
while True:
    try:
        num_dice_input = input("How many dice do you want to roll? ")
        num_dice = int(num_dice_input)
        if num_dice <= 0:
            print("Please enter a positive number of dice.")
            continue
    except ValueError:
        print("Invalid input. Please enter a number for the dice count.")
        continue

    dice_roller = input("Roll the dice? (y/n): ")
    if dice_roller.lower() == 'y':
        roll_count += 1
        rolls = []
        total = 0
        for _ in range(num_dice):
            die_roll = random.randint(1, 6)
            rolls.append(str(die_roll))
            total += die_roll
        print(f"You rolled {', '.join(rolls)}. Total: {total}")
        print(f"You have rolled the dice {roll_count} time(s) during this session.")
    elif dice_roller.lower() == 'n':
        print("Thank you for playing!")
        break
    else:
        print("Invalid choice! Please enter 'y' or 'n'.")
```
