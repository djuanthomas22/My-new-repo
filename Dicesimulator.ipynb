```python
import random
```

```python
# Ask: roll the dice?
while True:

    dice_roller = input("Roll the dice? (y/n): ")
    if dice_roller.lower() == 'y':
        # generate two random numbers for the dice
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
    if dice_roller.lower() == 'y':
        print(f"You rolled a {die1} and a {die2}. Total: {die1 + die2}")
    elif dice_roller.lower() == 'n':
        print("Thank you for playing!")
        break
    else: not 'y' or 'n'
    print("Invalid choice! Please enter 'y' or 'n'.")
```
